# Trabalho Mineração de Dados


A função `gera_tabela_info_dataset` tem como objetivo gerar uma tabela em formato markdown com informações sobre um dataset. Ela realiza as seguintes tarefas:

1. **Identificação do Tipo de Arquivo**:
   - Verifica se o arquivo é um CSV ou Excel.
   - Lê o arquivo correspondente utilizando `pd.read_csv` para arquivos CSV e `pd.read_excel` para arquivos Excel.
   - Lança um erro se o tipo de arquivo não for suportado.

2. **Criação da Tabela**:
   - Inicia uma string de tabela em markdown com cabeçalhos para "Variável", "Tipo" e "Valores Nulos".
   - Para cada coluna no dataset, identifica o tipo de dado (`dtype`) e o número de valores nulos, adicionando essas informações à tabela.

3. **Adição de Informações Adicionais**:
   - Adiciona o número de linhas e colunas do dataset ao final da tabela.

4. **Retorno da Tabela**:
   - Retorna a tabela em formato markdown como uma string.


````python

import pandas as pd

def gera_tabela_info_dataset(dataset):
  """
  Esta função recebe um dataset (csv ou xls) e retorna uma tabela em markdown com informações sobre o dataset.

  Args:
    dataset: Um objeto pandas.DataFrame.

  Returns:
    Uma string contendo a tabela em markdown.
  """

  # Identifica o tipo de arquivo
  if dataset.endswith('.csv'):
    df = pd.read_csv(dataset)
  elif dataset.endswith('.xls') or dataset.endswith('.xlsx'):
    df = pd.read_excel(dataset)
  else:
    raise ValueError('Tipo de arquivo não suportado.')

  # Cria a tabela
  tabela = f"| Variável | Tipo | Valores Nulos | \n|---|---|---| \n"

  # Adiciona as informações de cada variável
  for coluna in df.columns:
    tipo = df[coluna].dtype
    nulos = df[coluna].isnull().sum()
    tabela += f"| {coluna} | {tipo} | {nulos} | \n"

  # Adiciona o número de linhas e colunas
  tabela += f"\nNúmero de linhas: {df.shape[0]} \n"
  tabela += f"Número de colunas: {df.shape[1]}"

  return tabela,df

# Exemplo de uso
dataset = '1719048842869+default of credit card clients.xls'
tabela,df = gera_tabela_info_dataset(dataset)
print(tabela)
````

# Análise de agrupamentos

## Método K-means


````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Método do cotovelo para determinar o número ideal de clusters
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    sse.append(kmeans.inertia_)

# Plot do método do cotovelo
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('SSE')
plt.title('Método do Cotovelo')
plt.show()
````

O gráfico do método do cotovelo mostra a soma dos erros quadráticos dentro dos clusters (SSE) para diferentes números de clusters. O ponto ideal (o cotovelo) é onde a diminuição de SSE começa a se tornar menos acentuada. Isso indica que adicionar mais clusters não está proporcionando uma melhoria significativa na compactação dos clusters.

Observando o gráfico que você forneceu, parece que o cotovelo está em torno de 3 ou 4 clusters. Este é o ponto onde a curva começa a se "achatar", ou seja, a redução de SSE entre os clusters 3 e 4 é significativa, mas a redução adicional após 4 clusters se torna menos pronunciada.

Usaremos 3 clusters como o número ideal.

````python
# Aplicação do K-means com 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster_KMeans'] = kmeans.fit_predict(data_scaled)

# Análise dos clusters
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centers_df = pd.DataFrame(cluster_centers, columns=relevant_columns)
print("Centros dos Clusters (K-means):")
print(cluster_centers_df)

# Adicionalmente, podemos plotar os clusters para visualização
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=df['Cluster_KMeans'], cmap='viridis', marker='o', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.title('Clusters K-means')
plt.xlabel('Feature 1 (Normalizada)')
plt.ylabel('Feature 2 (Normalizada)')
plt.show()
````

## Método de Agrupamento Hierárquico


#### Determinação dos Parâmetros

1. **Linkage Method**:
   - Utilizamos o método `ward` que minimiza a variância dentro dos clusters ao unir pontos de dados.
   
2. **Número de Clusters (t)**:
   - O número de clusters foi determinado com base na análise visual do dendrograma. Escolhemos cortar o dendrograma para obter 3 clusters, o que nos proporciona uma boa divisão dos dados.


````python
from scipy.cluster.hierarchy import dendrogram, linkage

# Criação do dendrograma
linked = linkage(data_scaled, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Dendrograma')
plt.xlabel('Clientes')
plt.ylabel('Distância')
plt.show()
````

````python
from scipy.cluster.hierarchy import fcluster

# Determinação dos clusters
clusters_hierarchical = fcluster(linked, t=3, criterion='maxclust')
df['Cluster_Hierarchical'] = clusters_hierarchical

# Análise dos clusters
df['Cluster_Hierarchical'] = df['Cluster_Hierarchical'].astype('category')
hierarchical_summary = df.groupby('Cluster_Hierarchical')[relevant_columns].mean()
print("Resumo dos Clusters (Agrupamento Hierárquico):")
print(hierarchical_summary)
````

### Método DBSCAN

#### Determinação dos Parâmetros

1. **Escolha de $ \epsilon $**:
   - Para determinar $ \epsilon $ (epsilon), utilizamos o gráfico dos k-vizinhos mais próximos.
   - O gráfico mostra a distância ao vizinho mais próximo (eixo Y) para cada ponto de dados (eixo X).
   - O ponto de inflexão no gráfico é onde a curva começa a subir rapidamente, indicando a transição de pontos dentro dos clusters para pontos considerados ruído ou em regiões de baixa densidade.
   - A escolha do valor de $ \epsilon $ foi baseada no gráfico dos k-vizinhos mais próximos. O ponto de inflexão do gráfico indica onde a curva começa a subir rapidamente, que é o valor ideal para $ \epsilon $
   - Observando o gráfico, o ponto de inflexão ocorre em torno de $ \epsilon = 5 $ mas optamos por 3,  este valor foi escolhido para não usar tanta memória.

2. **`min_samples`**:
   - Um valor menor de `min_samples` ajuda a reduzir a quantidade de memória usada. Escolhemos `min_samples = 3` para balancear a formação de clusters com o uso de memória.



````python
from sklearn.neighbors import NearestNeighbors
import numpy as np

# Determinação de epsilon usando o gráfico dos k-vizinhos mais próximos
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(data_scaled)
distances, indices = neighbors_fit.kneighbors(data_scaled)

# Ordenação das distâncias e plotagem
distances = np.sort(distances, axis=0)
distances = distances[:, 1]
plt.plot(distances)
plt.xlabel('Pontos de Dados')
plt.ylabel('Distância')
plt.title('Gráfico dos k-vizinhos mais próximos')
plt.show()
````

````python
from sklearn.cluster import DBSCAN

# Aplicação do DBSCAN com valores menores de epsilon e min_samples
dbscan = DBSCAN(eps=3, min_samples=3)
clusters_dbscan = dbscan.fit_predict(data_scaled)
df['Cluster_DBSCAN'] = clusters_dbscan

# Análise dos clusters
unique, counts = np.unique(clusters_dbscan, return_counts=True)
print("Contagem de Clusters (DBSCAN):")
print(dict(zip(unique, counts)))

# Descrição dos clusters
dbscan_summary = df[df['Cluster_DBSCAN'] != -1].groupby('Cluster_DBSCAN')[relevant_columns].mean()
print("Resumo dos Clusters (DBSCAN):")
print(dbscan_summary)
````
